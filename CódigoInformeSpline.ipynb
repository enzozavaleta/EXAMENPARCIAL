```python
import numpy as np
from scipy.interpolate import CubicSpline

# 1. Datos experimentales discretos (Nodos de control)
frecuencias = np.array([100, 145, 235, 405, 730, 1000], dtype=float)
impedancias = np.array([152.3, 146.8, 139.5, 133.6, 130.9, 132.5], dtype=float)

# 2. Generar una malla fina continua para una reconstrucción suave
f_fina = np.linspace(100, 1000, 500)

# 3. Interpolación por Spline Cúbico Natural
spline_natural = CubicSpline(frecuencias, impedancias, bc_type='natural')
z_fina_spline = spline_natural(f_fina)

# 4. Evaluación de un punto de interés específico (f = 500 Hz)
f_interes = 500.0
z_estimada = spline_natural(f_interes)

# 5. Despliegue de resultados numéricos por consola
print('====================================================')
print('     RESULTADOS DEL PROCESAMIENTO POR SPLINES       ')
print('====================================================')
print(f'Frecuencia de control evaluada: {f_interes:10.2f} Hz')
print(f'Impedancia calculada por Spline: {z_estimada:10.4f} Ohm')
print('====================================================')
```
